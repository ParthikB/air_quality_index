```python
import pandas as pd
import numpy as np
from datetime import datetime
from tqdm import tqdm
```

```python
def convert(x):
    day, month, _ = list(map(int, x.split()[0].split('/')))
#     return datetime(YEAR, month, day).timestamp()
    return str(datetime(YEAR, month, day))
```

```python
df1 = pd.read_csv('database/raw_master_data.csv')
master = pd.DataFrame()

for YEAR in tqdm(range(2013, 2019)):
    df2 = pd.read_csv(f'database/AQI/aqi{YEAR}.csv')
    df2['Date'] = df2['Date'].apply(convert)

    col='PM2.5'
    for idx, val in enumerate(df2[col]):
        try:
            df2[col].iloc[idx] = float(df2[col].iloc[idx])
        except:
            df2[col].iloc[idx] = None

    mean_df = pd.DataFrame(df2.groupby('Date')[col].apply(lambda x: x.mean()))
    
#     master = pd.merge(left=master, right=mean_df, how='left', left_on='Day', right_on='Date')    
    merged_inner = pd.merge(left=df1, right=mean_df, left_on='Day', right_on='Date')
    
    master = pd.concat([master, merged_inner])
```

```python
master = master.set_index('Day')
master
```

## Exporting the CSV

```python
master.to_csv('database/raw_master_combined_data.csv')
```
